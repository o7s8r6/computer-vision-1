```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import cv2
%config InlineBackend.figure_format = "retina"
```

```python
image = cv2.cvtColor(cv2.imread('images/round_farms.jpg'),cv2.COLOR_BGR2RGB)
image_copy = np.copy(image)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
gaussian = cv2.GaussianBlur(gray,(3,3),0)
```

```python
plt.imshow(gaussian,cmap='gray')
```

```python
circle_im = image_copy
```

```python
circles = cv2.HoughCircles(gaussian, cv2.HOUGH_GRADIENT, 1, 
                           minDist=45,
                           param1=70,
                           param2=11,
                           minRadius=20,
                           maxRadius=40)

circles = np.uint16(np.around(circle_im))
for i in circles[0,:]:
    cv2.circle(circle_im,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(circle_im,(i[0],i[1]),2,(0,0,255),3)
    
plt.imshow(circle_im)
```
