### Canny Edge Detection 
`Widely used and accurate edge detection algorithm`
1. Filters out noise using Gaussian Blur
2. Finds the strength and direction of edges using Sobel filters
3. Applies non-maximum suppression to isolate the strongest edges and 
    thin them to one-pixel wide lines
4. Uses hysteresis to isolate the best edges

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
%config InlineBackend.figure_format = "retina"
```

```python
image = cv2.imread('images/sunflower.jpg')
img_copy = np.copy(image)
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)
plt.imshow(img_copy,cmap='gray')
```

```python
gb = cv2.GaussianBlur(img_copy,(5,5),0)
plt.imshow(gb, cmap='gray')
```

```python
sobel = np.array([[-1,0,1],
         [-2,0,2],
         [-1,0,1]])
```

```python
filtererd_img = cv2.filter2D(gb,-1,sobel)
```

```python
plt.imshow(filtererd_img,cmap='gray')
```

```python
lower = 120
upper = 240
edges = cv2.Canny(filtererd_img,lower,upper)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(edges,cmap='gray')
```
